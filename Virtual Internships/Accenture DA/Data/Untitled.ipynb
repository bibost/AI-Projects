### Data Cleaning & Exploration Accenture - Social Buzz

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data Loading

content = pd.read_csv("Content.csv")
reactions = pd.read_csv("Reactions.csv")
reaction_types = pd.read_csv("ReactionTypes.csv")
```

```python
content.isna().sum()
```

```python
reactions.isna().sum()
```

```python
reaction_types.isna().sum()
```

```python
content = content.dropna()
```

```python
reactions = reactions.dropna()
```

```python
content = content.drop("User ID", axis=1)
content = content.drop("Unnamed: 0", axis=1)
content = content.drop("URL", axis=1)

content
```

```python
reactions = reactions.drop("User ID", axis=1)
reactions = reactions.drop("Datetime", axis=1)
reactions = reactions.drop("Unnamed: 0", axis=1)

reactions
```

```python
reaction_types = reaction_types.drop("Sentiment", axis=1)
reaction_types = reaction_types.drop("Unnamed: 0", axis=1)

reaction_types
```

```python
reactions_content = pd.merge(reactions, content, on="Content ID")
```

```python
reactions_content
```

```python
reaction_types.rename(columns={'Type': 'Type_x'}, inplace=True)
```

```python
reactions_content_types = pd.merge(reactions_content, reaction_types, on="Type_x")
```

```python
reactions_content_types
```

```python
result_df = reactions_content_types.groupby(['Content ID', 'Category'])['Score'].sum().reset_index()
```

```python
result_df
```

```python
result_df = result_df.sort_values(by="Score", ascending=False)
```

```python
result_df
```

```python
result_df.to_csv('final_dataset.csv', index=False)
```
