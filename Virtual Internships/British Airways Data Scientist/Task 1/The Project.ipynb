# British Airways Project

### Problem

* Find out what customers think about British Airways using various ML & DA Methods.

### The Data

* The data we are working with is a dataset scraped from Skytrax using BeautifulSoup
* It is a wide variety of customer reviews stacked into a CSV file

### Goal

* Pre-Process The Data
* Create WordCloud for Corpus
* LDA & WordClouds
* Sentiment Analysis

### Importing Packages

```python
import pandas as pd
import numpy as np
import re
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import gensim
from gensim import corpora, models
import warnings
warnings.filterwarnings('ignore')
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('vader_lexicon')
```

### Loading & Pre-Processing The Data

```python
#Loading Data

data = pd.read_csv("data/BA_reviews.csv")
```

```python
data
```

```python
#Removing Unnecessary Column

data = data.drop('Unnamed: 0', axis=1)
```

```python
#Removing Unnecessary Prefixes

data['reviews'] = data['reviews'].str.replace(r'^(✅ Trip Verified | Not verified )', '', regex=True)
```

```python
data
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

#Creating a pre-processing function

def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation and numbers
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize and remove stopwords
    tokens = text.split()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)
```

```python
data['reviews'] = data['reviews'].apply(preprocess_text) #Applying the pre-processing
```

```python
data #Final Pre-Processed Data Set
```

# Generating a World Cloud from total reviews

```python
text_corpus = ' '.join(data['reviews']) #Creating a text corpus

# Create and generate a word cloud image
wordcloud = WordCloud(background_color='white').generate(text_corpus)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.savefig('wordcloud_total.png')
plt.show()
```

# LDA & Word Clouds from LDA Topics

```python
def preprocess_text_LDA(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation and numbers
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize and remove stopwords
    tokens = text.split()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return tokens  # Return list of tokens instead of a string
```

```python
data['lda_reviews'] = data['reviews'].apply(preprocess_text_LDA) #Applying the pre-processing
```

```python
dictionary = corpora.Dictionary(data['lda_reviews'])

# Filter out extremes to remove tokens that appear too frequently or too rarely
dictionary.filter_extremes(no_below=5, no_above=0.5)

# Convert document into the bag-of-words (BoW) format = list of (token_id, token_count) tuples
corpus = [dictionary.doc2bow(text) for text in data['lda_reviews']]
```

```python
lda_model = models.LdaModel(corpus=corpus, num_topics=10, id2word=dictionary, passes=15)

# View the topics
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```

```python
for t in range(lda_model.num_topics):
    plt.figure()
    plt.imshow(WordCloud(background_color='white').fit_words(dict(lda_model.show_topic(t, 20))))
    plt.title('Topic ' + str(t))
    plt.axis("off")
    plt.savefig(f'wordcloud{t}.png')
    plt.show()
```

# Sentiment Analysis using VADER

```python
# Initializing VADER
sid = SentimentIntensityAnalyzer()

def get_vader_compound(text):
    scores = sid.polarity_scores(text)
    return scores['compound']

data['compound'] = data['reviews'].apply(get_vader_compound)

# Classifying sentiments as positive, neutral, or negative based on the compound score
data['sentiment'] = data['compound'].apply(lambda x: 'positive' if x > 0.05 else ('negative' if x < -0.05 else 'neutral'))
```

```python
data_sentiment = data['sentiment'].value_counts()
```

```python
data_sentiment = pd.DataFrame(data_sentiment)
```

# LDA WordClouds for Positive & Negative Sentiments

```python
data_lda_positive = data[data['sentiment'] == 'positive']
```

```python
data_lda_negative = data[data['sentiment'] == 'negative']
```

```python
dictionary_positive = corpora.Dictionary(data_lda_positive['lda_reviews'])

dictionary_positive.filter_extremes(no_below=5, no_above=0.5)

corpus_positive = [dictionary_positive.doc2bow(text) for text in data_lda_positive['lda_reviews']]
```

```python
lda_model = models.LdaModel(corpus=corpus_positive, num_topics=3, id2word=dictionary_positive, passes=15)

for t in range(lda_model.num_topics):
    plt.figure()
    plt.imshow(WordCloud(background_color='white').fit_words(dict(lda_model.show_topic(t, 20))))
    plt.title('Topic ' + str(t))
    plt.axis("off")
    plt.savefig(f'wordcloud_positive{t}.png')
    plt.show()
```

```python
dictionary_negative = corpora.Dictionary(data_lda_negative['lda_reviews'])

dictionary_negative.filter_extremes(no_below=5, no_above=0.5)

corpus_negative = [dictionary_negative.doc2bow(text) for text in data_lda_negative['lda_reviews']]
```

```python
lda_model = models.LdaModel(corpus=corpus_negative, num_topics=3, id2word=dictionary_negative, passes=15)

for t in range(lda_model.num_topics):
    plt.figure()
    plt.imshow(WordCloud(background_color='white').fit_words(dict(lda_model.show_topic(t, 20))))
    plt.title('Topic ' + str(t))
    plt.axis("off")
    plt.savefig(f'wordcloud_negative{t}.png')
    plt.show()
```
