# Data Modelling

> The idea is to predict revenue generated in Dollars $USD per county for the following calendar year

## Pre-processing & Pipeline Building

```python
display(data)
```

### Feature Engineering

```python
data.info()
```

> I've chosen the following features as others do not make sense on a county level & others are leaking data

```python
modelling_columns = ['Date', 'County Number', 'County', 'Pack', 'Bottle Volume (ml)', 'State Bottle Cost', 'State Bottle Retail', 'Sale (Dollars)']
```

```python
data_modelling = data[modelling_columns]
```

```python
data_modelling.head()
```

```python
data_modelling['Date'] = data_modelling['Date'].dt.year
```

```python
data_modelling
```

```python
data_modelling['State Bottle Cost'] = data_modelling['State Bottle Cost'].str.replace('$', '').astype(float)
```

```python
data_modelling['State Bottle Retail'] = data_modelling['State Bottle Retail'].str.replace('$', '').astype(float)
```

```python
county_modelling = data_modelling.groupby(['County Number', 'Date']).agg({'Sale (Dollars)': 'sum', 'Pack': 'mean', 'Bottle Volume (ml)': 'mean', 'State Bottle Cost': 'mean', 'State Bottle Retail': 'mean'}).reset_index()
```

```python
county_modelling
```

### Data Splitting & Training

```python
X = county_modelling.drop('Sale (Dollars)', axis=1)
y = county_modelling['Sale (Dollars)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

### Model Evaluation

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
mse
```

```python
r2
```

```python
scores = cross_val_score(model, X, y,cv=5, scoring='r2')
```

```python
scores.mean()
```

> The model is not perfoming the best, probably because there are many counties and not enough historical entries for each, nor are the features the best.
